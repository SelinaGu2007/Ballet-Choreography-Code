```python
import pandas as pd
import math

# Load the Excel file
data = pd.read_excel(r'C:\Users\Gusul\Desktop\cleaned_head_coordinates_3d_2.xlsx')

# Create keypoint coordinates as tuples
marked_points = list(zip(data.iloc[:, 0], data.iloc[:, 1], data.iloc[:, 2]))

# Euclidean distance calculation function
def euclidean_distance(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[2] - p2[2])**2)

def period_energy_expenditure (start_time, end_time):
    
    # Calculate energy expenditure in the x-z plane
    distance_xz = 0
    for i in range (start_time, end_time - 1):
        distance_xz += euclidean_distance(marked_points[i], marked_points[i+1])
    W_xz = distance_xz * 0.4 * 9.8

    # Calculate energy expenditure in the y-axis
    distance_up = 0
    for i in range (start_time, end_time - 1):
        if marked_points[i][1] < marked_points[i+1][1]:
            distance_up += (marked_points[i+1][1] - marked_points[i][1])
    W_up = distance_up * 9.8

    # Calculate total energy expenditure
    W = W_xz + W_up
    return W

period_length = len(marked_points) // 5  # Calculate the length of each period

# Initialize an empty list to store energy values
energy_values = []

# Calculate energy expenditure in each of the 5 periods
variance = 0
for i in range(5):
    energy_period = 0
    start_index = i * period_length
    end_index = start_index + period_length
    energy = period_energy_expenditure (start_index, end_index)  
    print(f"Energy expenditure for period {i+1}: {energy}")
    
    # Append the calculated energy to the list
    energy_values.append(energy)

# Convert the list to a pandas Series
data_series = pd.Series(energy_values)

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = data_series.quantile(0.25)
Q3 = data_series.quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
low_outliers = data_series[(data_series < lower_bound)]
high_outliers = data_series[(data_series > upper_bound)]

print (f"Upper bound: {upper_bound}")
print (f"Lower bound: {lower_bound}")

# Check if there are low outliers
if not low_outliers.empty:
    print("Energy expenditure too low:")
    for value in low_outliers:
        print(value)

# Check if there are high outliers
if not high_outliers.empty:
    print("Energy expenditure too high:")
    for value in high_outliers:
        print(value)
```
